```python
%matplotlib inline
# Dependencies and Setup
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np

# File to Load (Remember to change these)
city_data_to_load = "data/city_data.csv"
ride_data_to_load = "data/ride_data.csv"

# Read the City and Ride Data
city_data = pd.read_csv(city_data_to_load)
ride_data = pd.read_csv(ride_data_to_load)

# Combine the data into a single dataset
#pd.merge(left, right, on='key')
combined_data = pd.merge(city_data, ride_data, on='city')

# Display the data table for preview
combined_data.head()
```

```python
# Get the total number of rides per city
rides_per_city = combined_data.groupby('city')['ride_id'].count()
rides_per_city.head()
```

```python
# Get the average fare per city
average_fair_per_city = combined_data.groupby('city')['fare'].mean()
average_fair_per_city.head()
```

```python
# Get the total number of drivers per city
drivers_per_city = city_data.set_index('city')['driver_count']
drivers_per_city.head()
```

```python
# Get the list of cities and city type
city_type = city_data.set_index('city')['type']
city_type.head()
```

```python
# Create a data frame to hold all the information
pyber_cities_df = pd.DataFrame({"Average Fare ($) Per City": average_fair_per_city,
                             "Total Number of Rides Per City": rides_per_city,
                             "Total Number of Drives Per City": drivers_per_city,
                             "City Type (Urban, Suburban, Rural)": city_type})
pyber_cities_df.head()
```

```python
# Get Urban DataFrame
urban_df = pyber_cities_df.loc[pyber_cities_df['City Type (Urban, Suburban, Rural)'] == 'Urban']

# Get Suburban DataFrame
suburban_df = pyber_cities_df.loc[pyber_cities_df['City Type (Urban, Suburban, Rural)'] == 'Suburban']

#Get Rural DataFrame
rural_df = pyber_cities_df.loc[pyber_cities_df['City Type (Urban, Suburban, Rural)'] == 'Rural']
```

## Bubble Plot of Ride Sharing Data

```python
# Obtain the x and y coordinates for each of the three city types
urban_x = urban_df['Total Number of Rides Per City']
urban_y = urban_df['Average Fare ($) Per City']

suburban_x = suburban_df['Total Number of Rides Per City']
suburban_y = suburban_df['Average Fare ($) Per City']

rural_x = rural_df['Total Number of Rides Per City']
rural_y = rural_df['Average Fare ($) Per City']

# Build the scatter plots for each city types
plt.scatter(urban_x, urban_y, s=urban_df['Total Number of Drives Per City']*10, c='lightcoral', 
            alpha=0.75, linewidths= 1, edgecolors='black', label='Urban')

plt.scatter(suburban_x, suburban_y, s=suburban_df['Total Number of Drives Per City']*10, c='lightskyblue', 
            alpha=0.75, linewidths=1, edgecolors='black', label='Suburban')

plt.scatter(rural_x, rural_y, s=rural_df['Total Number of Drives Per City']*10, c='gold', 
            alpha=0.75, linewidths=1, edgecolors='black', label='Rural')

# Incorporate the other graph properties
plt.title('Pyber Ride Sharing Data (2016)')
plt.xlabel('Total Number of Rides (Per City)')
plt.ylabel('Average Fare ($)')
plt.text(42, 35, 'Note: \n Circle size correlates with driver count per city.')

# Create a legend
legend = plt.legend(loc='upper right', title='City Types')

# Incorporate a text label regarding circle size
legend.legendHandles[0]._sizes = [30]
legend.legendHandles[1]._sizes = [30]
legend.legendHandles[2]._sizes = [30]

plt.grid()

# Save Figure
plt.savefig('PyberRideData2016.png')

# Show plot
plt.show()
```

## Total Fares by City Type

```python
# Calculate Type Percents
fares_by_city_type = combined_data.groupby('type')['fare'].sum()

# Build Pie Chart
labels = ['Rural', 'Suburban', 'Urban']
colors = ['gold', 'lightskyblue', 'lightcoral']
plt.pie(fares_by_city_type, labels=labels, colors=colors, explode=(0, 0, 0.1),  
        shadow=True, startangle=150, autopct='%1.1f%%')
plt.title('% of Total Fares by City Type')
plt.axis('equal')
plt.tight_layout()

# Save Figure
plt.savefig('FaresbyCityType.png')

# Show Figure
plt.show()
```

## Total Rides by City Type

```python
# Calculate Ride Percents
rides_by_city_type = combined_data.groupby('type')['ride_id'].count()

# Build Pie Chart
labels = ['Rural', 'Suburban', 'Urban']
colors = ['gold', 'lightskyblue', 'lightcoral']
plt.pie(rides_by_city_type, labels=labels, colors=colors, explode=(0, 0, 0.1),  
        shadow=True, startangle=150, autopct='%1.1f%%')
plt.title('% of Total Rides by City Type')
plt.axis('equal')
plt.tight_layout()

# Save Figure
plt.savefig('RidesbyCityType.png')

# Show Figure
plt.show()
```

## Total Drivers by City Type

```python
# Calculate Driver Percents
drivers_by_city_type = city_data.groupby('type')['driver_count'].sum()

# Build Pie Chart
labels = ['Rural', 'Suburban', 'Urban']
colors = ['gold', 'lightskyblue', 'lightcoral']
plt.pie(drivers_by_city_type, labels=labels, colors=colors, explode=(0, 0, 0.1),  
        shadow=True, startangle=150, autopct='%1.1f%%')
plt.title('% of Total Drives by City Type')
plt.axis('equal')
plt.tight_layout()

# Save Figure
plt.savefig('DriversbyCityType.png')

# Show Figure
plt.show()
```
