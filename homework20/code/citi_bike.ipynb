## Citi Bike Daily Ridership and Membership Data
### 2019 Data: Jan - Mar; Apr - Jun

```python
import pandas as pd
```

```python
q1df = pd.read_csv("../csv_files/citiBike2019Q1.csv") 
q2df = pd.read_csv("../csv_files/citiBike2019Q2.csv")
```

```python
q1df.head()
```

```python
q2df.head()
```

```python
df = q1df.append(q2df, sort=False)
```

```python
df.head()
```

```python
df.count()
```

```python
df = df[["Date", "Trips over the past 24-hours (midnight to 11:59pm)", 
         "Miles traveled today (midnight to 11:59 pm)", 
         "Total Annual Members (All Time)", 
         "24-Hour Passes Purchased (midnight to 11:59 pm)", 
         "3-Day Passes Purchased (midnight to 11:59 pm)"]]
```

```python
df.head()
```

```python
df.count()
```

```python
df.rename(columns={'Trips over the past 24-hours (midnight to 11:59pm)':'TotalTrips',
                  'Miles traveled today (midnight to 11:59 pm)':'TotalTraveledMiles',
                  'Total Annual Members (All Time)':'TotalMembers',
                  '24-Hour Passes Purchased (midnight to 11:59 pm)':'Total24hrPasses',
                  '3-Day Passes Purchased (midnight to 11:59 pm)':'Total3DayPasses'},
         inplace=True)
```

```python
df.head()
```

```python
citiBikeData = df.to_csv("../csv_files/2019CitiBikeData.csv", index = None, header=True)
```

### Combine all the trips data

```python
tripdata01 = pd.read_csv("../csv_files/JC-201901-citibike-tripdata.csv")
tripdata02 = pd.read_csv("../csv_files/JC-201902-citibike-tripdata.csv")
tripdata03 = pd.read_csv("../csv_files/JC-201903-citibike-tripdata.csv")
tripdata04 = pd.read_csv("../csv_files/JC-201904-citibike-tripdata.csv")
tripdata05 = pd.read_csv("../csv_files/JC-201905-citibike-tripdata.csv")
tripdata06 = pd.read_csv("../csv_files/JC-201906-citibike-tripdata.csv")
```

```python
tripdata = pd.concat([tripdata01,tripdata02,tripdata03,tripdata04,tripdata05,tripdata06])
```

```python
tripdata.dtypes
```

```python
tripdata.gender.replace([0,1,2], ['unknown','male','female'], inplace=True)
```

```python
tripdata.head()
```

```python
bikeTripData = tripdata.to_csv("../csv_files/2019CitiBikeTripData.csv", index = None, header=True)
```
